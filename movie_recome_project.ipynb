```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):#string of list of dictionaries to lists
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
def convert3(obj):#to get 3 outer names
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
          L.append(i['name'])
          counter+=1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):#fetching  each movie director
    L=[]
    for i in ast.literal_eval(obj):
         if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]

    distances=similarity[movie_index]

    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]#The list is sorted in descending order based on similarity scores

    for i in movies_list:
        print (new_df.iloc[i[0]].title)

```
